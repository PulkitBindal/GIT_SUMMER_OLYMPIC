### 1. In how many cities Summer Olympics is held so far?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("summer.csv")

data = []

for city in df['City'].unique():
  data.append([city , len(df[df['City']  == city])])
print("The Summer Olympics have been held in"+ " "+str(len(data))+" " +"countries so far.\n")
print("The graphical representation is shown below :\n")
pd.DataFrame(data,columns = ['city','Frequency']).sort_values(by='Frequency', ascending=False).plot(x = 'city', y = 'Frequency', kind = 'bar', figsize = (15,5), color = 'orange')
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("summer.csv")

data = []

for Sport in df['Sport'].unique():
    data.append([Sport , len(df[df['Sport']  == Sport])])
    
print("The graphical representation is shown below :\n")    
pd.DataFrame(data,columns = ['Sport','Most Medals']).sort_values(by='Most Medals', ascending=False).head().plot(x = 'Sport', y = 'Most Medals', kind = 'bar', figsize = (5,5), color = 'pink')
```
